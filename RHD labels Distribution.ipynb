```python
import pandas as pd
import matplotlib.pyplot as plt

labels = pd.read_csv("Cloud_SQL_Export_2021-09-24 (13_40_21) - Sorted.csv",index_col=0)

labels.info()
```

```python
labels.head()
```

```python
figure, axes = plt.subplots(figsize=(11,11),nrows=3, ncols=3)

labels["VIEW"].value_counts().plot(ax=axes[0,0], kind='barh', legend=True, grid=True, color='indigo', rot=0, width=0.2)
labels["COLOUR"].value_counts().plot(ax=axes[0,1], kind='barh', legend=True, grid=True, color='red', rot=0, width=0.1)
labels["VALIDATION"].value_counts().plot(ax=axes[0,2], kind='barh', legend=True, grid=True, color='magenta', rot=0, width=0.1)
labels["VIEW-APP"].value_counts().plot(ax=axes[1,0], kind='barh', legend=True, grid=True, color='cyan', rot=0, width=0.3)
labels["THICKNESS"].value_counts().plot(ax=axes[1,1], kind='barh', legend=True, grid=True, color='green', rot=0, width=0.2)
labels["CONDITION"].value_counts().plot(ax=axes[2,1], kind='barh', legend=True, grid=True, color='blue', rot=0, width=0.35)
labels["SEVERITY"].value_counts().plot(ax=axes[1,2], kind='barh', legend=True, grid=True, color='purple', rot=0, width=0.25)

figure.suptitle("Distribution of labelled RHD Data", fontsize=18)
figure.subplots_adjust(top=0.95)

figure.subplots_adjust(hspace=0.2, wspace=0.8)
figure.delaxes(axes[2][0])
figure.delaxes(axes[2][2])

figure.savefig("labels_distribution.jpg",
              dpi=400,
             bbox_inches='tight', 
             pad_inches = 0.2,
             transparent = True,
             facecolor ="w",
             edgecolor ='w',
             orientation ='landscape')
plt.show()
```
